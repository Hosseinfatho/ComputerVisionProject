```python
import os
import shutil
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array, array_to_img
from skimage.util import random_noise

# Paths
BASE_DIR = "/content/drive/MyDrive/CS415Project/chest_xray"
ORIGINAL_FOLDER = os.path.join(BASE_DIR, "Original")
NOISY_FOLDER = os.path.join(BASE_DIR, "Noisy")
PROCESSED_TEST_FOLDER = os.path.join(BASE_DIR, "processed_test")

# Function to add random noise
def add_random_noise(image_array):
    return random_noise(image_array, mode='gaussian', clip=True)

# Step 1: Move and rename images to "Original"
def create_original_folder():
    # Create the "Original" folder
    if not os.path.exists(ORIGINAL_FOLDER):
        os.makedirs(ORIGINAL_FOLDER)

    # Process each category folder
    categories = ['BACTERIA', 'NORMAL', 'VIRUS']
    image_counter = 1

    for category in categories:
        category_path = os.path.join(PROCESSED_TEST_FOLDER, category)

        for image_name in os.listdir(category_path):
            if image_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                original_image_path = os.path.join(category_path, image_name)
                new_image_name = f"{image_counter}.jpg"
                new_image_path = os.path.join(ORIGINAL_FOLDER, new_image_name)

                # Move and rename the image
                shutil.copy(original_image_path, new_image_path)
                print(f"Moved and renamed: {original_image_path} -> {new_image_path}")
                image_counter += 1

# Step 2: Create "Noisy" folder with random noise
def create_noisy_folder():
    # Create the "Noisy" folder
    if not os.path.exists(NOISY_FOLDER):
        os.makedirs(NOISY_FOLDER)

    # Process each image in the "Original" folder
    for image_name in os.listdir(ORIGINAL_FOLDER):
        if image_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            original_image_path = os.path.join(ORIGINAL_FOLDER, image_name)
            noisy_image_path = os.path.join(NOISY_FOLDER, image_name)

            # Load and process the image
            image = load_img(original_image_path)
            image_array = img_to_array(image) / 255.0  # Normalize to [0, 1]
            noisy_image_array = add_random_noise(image_array)

            # Convert back to image and save
            noisy_image = array_to_img(noisy_image_array)
            noisy_image.save(noisy_image_path)
            print(f"Saved noisy image: {noisy_image_path}")

# Execute steps
create_original_folder()
create_noisy_folder()
```
